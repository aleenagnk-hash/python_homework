#What range() does

range() produces a sequence of numbers — but it’s not a list.
Instead, it’s a special object that generates numbers one by one when needed (this saves memory).

```python
print('a_range_is_not_a_list', (range(0-3)))
```

```python
for numbers in range (0,3):
    print (numbers)
```

A for loop has three parts:

Collection	The sequence you’re looping over	
Loop variable	The temporary variable that takes each value	
Body	The indented block that runs every time	

```python
for numbers in range (10):
    total = total+ (numbers+1)
    print (total)
```

```python
total = 0
for number in range(10):
    total = total + (number + 1)
print(total)
```

Version 1 → shows progress step by step (inside loop).
Version 2 → shows only the final result (after loop).
