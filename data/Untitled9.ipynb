```python
| Step | `char` (current letter) | Previous `result` | Operation `result = char + result` | New `result` |
| ---- | ----------------------- | ----------------- | ---------------------------------- | ------------ |
| 1    | `"t"`                   | `""`              | `"t" + ""`                         | `"t"`        |
| 2    | `"i"`                   | `"t"`             | `"i" + "t"`                        | `"it"`       |
| 3    | `"n"`                   | `"it"`            | `"n" + "it"`                       | `"nit"`      |
```

##### Create an acronym: Starting from the list ["red", "green", "blue"], create the acronym "RGB" using a for loop.

Hint: You may need to use a string method to properly format the acronym.

```python
acronym = ""
for word in ["red", "green", "blue"]:
    acronym = acronym + word[0].upper()
print(acronym)
```

```python
| Loop | word  | word[0].upper() | New acronym |
| ---- | ----- | --------------- | ----------- |
| 1    | red   | R               | `"R"`       |
| 2    | green | G               | `"RG"`      |
| 3    | blue  | B               | `"RGB"`     |
```

###### acronym = acronym + word[0].upper() - This line takes the first letter of each word, makes it uppercase, and keeps adding it to the acronym string step by step.
